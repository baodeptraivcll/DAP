```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Tạo dữ liệu giả định
area = np.array([50, 70, 80, 100, 120]) # Diện tích của ngôi nhà (m2)
price = np.array([200, 270, 300, 370, 450]) # Giá của ngôi nhà (nghìn USD)
```

```python
area = area.reshape(-1, 1)
X = area
y = price
```

```python
model = LinearRegression()
model.fit(X, y)
pred = model.predict(X)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, color = 'Blue', label = 'Actual Price')
plt.plot(X, pred, color = 'Red', label = 'Predicted Price')
plt.title('House Price Prediction')
plt.xlabel('Area (m2)')
plt.ylabel('Price (1000 USD)')
plt.legend()
plt.show()
```
