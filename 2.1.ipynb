```python
import pandas as pd
import statsmodels.api as sm
```

```python
url = 'https://raw.githubusercontent.com/lenhattung/SP26-AI1904-DAP391m/refs/heads/main/lab_part_2/01%20-%20Linear%20and%20logistic%20regression/wines.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.quality.unique()
```

```python
y = df['quality']
X = df[['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
                                'chlorides', 'free_sulfur_dioxide',
                                'total_sulfur_dioxide', 'density', 'pH',
                                'sulphates', 'alcohol']]
X = sm.add_constant(X)
model = sm.OLS(y, X)
result= model.fit()
result.summary()
```
