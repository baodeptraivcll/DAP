```python
import os
import cv2

# --- HÀM CẮT ẢNH GỌN NHẸ ---
# SỬA 1: Đổi chỗ video_count lên trước frames_skip để không bị lỗi Syntax
def video_2_image(input_path, output_folder, video_count, frames_skip=40):

    # Chỉ giữ lại mỗi lệnh tạo folder này thôi (cần thiết)
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    video = cv2.VideoCapture(input_path)

    frame_curr = 0 # Đếm thời gian trong video này
    save_curr = video_count # Lấy số đếm tiếp nối từ video trước

    print(f"▶️ Đang chạy: {os.path.basename(input_path)} | Bắt đầu số: {save_curr}")

    while True:
        success, frame = video.read()
        if not success:
            break

        if frame_curr % frames_skip == 0:
            file_name = f'bien_so_{save_curr:04d}.jpg'
            file_path = os.path.join(output_folder, file_name)

            cv2.imwrite(file_path, frame)

            # In ra cho đỡ buồn ngủ, 10 tấm in 1 lần
            if save_curr % 10 == 0:
                print(f'   -> Đã lưu: {file_name}')

            save_curr += 1 # Tăng số thứ tự ảnh

        frame_curr += 1

    video.release()
    # Trả về con số cuối cùng để video sau đếm tiếp
    return save_curr

# --- PHẦN CHẠY CHÍNH ---

input_folder = '/content/drive/MyDrive/bike_plates/raw_vids/FPTU'
output_folder = '/content/drive/MyDrive/bike_plates/raw_vids/FPTU/anh_FPTU'

# Lấy danh sách file (Bạn đảm bảo 100% là video thì mình không lọc nữa)
list_files = os.listdir(input_folder)
list_files.sort() # Sắp xếp tên video chạy cho đúng thứ tự 1, 2, 3...

# Biến đếm tổng (Global)
global_count = 0

for file_name in list_files:
    # Bỏ qua nếu lỡ code đọc trúng cái folder output 'anh_FPTU' nằm bên trong
    if file_name == 'anh_FPTU':
        continue

    full_path = os.path.join(input_folder, file_name)

    # SỬA 2: Cập nhật biến global_count sau khi chạy xong hàm
    # Để video sau biết đường mà đếm tiếp
    global_count = video_2_image(full_path, output_folder, global_count, frames_skip=40)

print(f"✅ XONG! Tổng cộng: {global_count} ảnh.")



```
