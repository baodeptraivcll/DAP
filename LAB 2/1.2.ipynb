```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/lenhattung/SP26-AI1904-DAP391m/refs/heads/main/lab_part_2/01%20-%20Linear%20and%20logistic%20regression/titanic.csv'
df = pd.read_csv(url)
```

```python
embarked_dummies = pd.get_dummies(df["embarked"], prefix="embarked", drop_first=True)
df = pd.concat([df, embarked_dummies], axis=1)
df = df.drop("embarked", axis=1)
df.head()
```

```python
a = {0 : 'Male', 1 : 'Female'}
df['sex'] = df['sex'].map(a)
print(df['survived'].unique())
print(df['survived'].value_counts())
```

```python
a = {0 : 'No', 1 : 'Yes'}
df['survived'] = df['survived'].map(a)
print(df['survived'].unique())
print(df['survived'].value_counts())
```

```python
titanic = pd.read_csv(url)
a = {1 : '1st', 2 : '2nd', 3 : '3rd'}
titanic['pclass'] = titanic['pclass'].map(a)
titanic['pclass'] = pd.Categorical(titanic['pclass'], categories = ['1st', '2nd', '3rd'], ordered = True)
print (titanic['pclass'].unique())
print (titanic['pclass'].value_counts())
```
