```python
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
url = 'https://raw.githubusercontent.com/lenhattung/SP26-AI1904-DAP391m/refs/heads/main/lab_part_2/01%20-%20Linear%20and%20logistic%20regression/wines.csv'
df = pd.read_csv(url)
df.info()
```

```python
# 2. Xác định biến phụ thuộc (Y) và độc lập (X)
y = df['quality']
X = df.drop(['quality', 'type'], axis=1)

# 3. Thêm hằng số (Intercept)
# Bắt buộc để mô hình có hệ số chặn
X = sm.add_constant(X)

# 4. Xây dựng mô hình Multinomial Logistic Regression
# MNLogit tự động hiểu y là các nhãn (categories) riêng biệt
model = sm.MNLogit(y, X)

# 5. Huấn luyện mô hình
# method='bfgs' thường giúp hội tụ tốt hơn với dữ liệu phức tạp
results = model.fit(method='bfgs', maxiter=10000) 

# 6. Xem kết quả
print(results.summary())
```
